```python
import plotly.offline as pyo
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
```

```python
import pandas as pd
df_kghm = pd.read_csv('kgh_d.csv')
df_kghm.head()
```

```python
df_ceny_miedzi = pd.read_csv('ca_c_f_d.csv')
df_ceny_miedzi.head()
```

Stwórz dwa osobne wykresy liniowe (jeden pod drugim) – jeden niech przedstawia ceny zamknięcia KGHM, drugi niech przedstawia ceny miedzi (na podstawie kolumn "Zamknięcie").

```python
import plotly.express as px
```

```python
fig = px.line(df_kghm, 
                 x='Data', 
                 y='Zamkniecie', 
                 title='KGHM')
fig.update_traces(
    line=dict(
        color="blue",  # Set line color (can be a named color, hex code, or RGB)
        width=3,       # Set line width
        dash="solid",  # Set line style ("solid", "dash", "dot", "dashdot")
        shape="linear" # Set line shape ("linear", "spline")
    )
)
fig.update_layout(
    xaxis_title="Rozwoj na przestrzeni lat",  # X-axis label
    yaxis_title="Cena zamkniecia akcji",  # Y-axis label
    title_x=0.5,
    )
fig.show()
```

```python
fig = px.line(df_ceny_miedzi, 
                 x='Data', 
                 y='Zamkniecie', 
                 title='Miedz')
fig.update_traces(
    line=dict(
        color="red",  # Set line color (can be a named color, hex code, or RGB)
        width=3,       # Set line width
        dash="solid",  # Set line style ("solid", "dash", "dot", "dashdot")
        shape="linear" # Set line shape ("linear", "spline")
    )
)
fig.update_layout(
    xaxis_title="Rozwoj na przestrzeni lat",  # X-axis label
    yaxis_title="Cena zamkniecia akcji",  # Y-axis label
    title_x=0.5,
    )
fig.show()
```

Jako trzeci element wykresu dodaj tabelę zestawiającą ceny zamknięcia KGHM oraz miedzi w kolejnych dniach (ponownie na podstawie kolumny "Zamknięcie").

```python
merged_inner = pd.merge(df_kghm, df_ceny_miedzi, on='Data')
```

```python
merged_inner.head()
```

```python
merged_inner = merged_inner[['Data','Zamkniecie_x','Zamkniecie_y']]
```

```python
merged_inner.head()
```

```python
merged_inner = merged_inner.rename(columns={'Zamkniecie_x': 'KGHM','Zamkniecie_y': 'Miedz'})
```

```python
print(merged_inner)
```

```python
merged_inner.head()
```

```python
from dash import Dash, html, dash_table
```

```python
app = Dash(__name__)

app.layout = html.Div([  # Corrected assignment
    dash_table.DataTable(
        data=merged_inner.to_dict('records'), # Convert to dictionary in this format [
#    {'KGHM': 102.4, 'Miedz': 6309.0},  # First row
#    {'KGHM': 100.09, 'Miedz': 6216.0}, # Second row
#    {'KGHM': 104.89, 'Miedz': 6170.0}  # Third row]
        columns=[
            {"name": "Data", "id": "Data"},
            {"name": "KGHM", "id": "KGHM"},  
            {"name": "Miedz", "id": "Miedz"}  
        ],
        style_table={'overflowX': 'auto'},
        style_cell={'textAlign': 'left'},
        style_data_conditional=[
            {
                'if': {'row_index': 'odd'},
                'backgroundColor': 'rgb(220, 220, 220)'
            }
        ]
    )
])

app.run_server(debug=True)
```

Polecenie z Kodilli: Wykresowi nadaj odpowiednie tytuły, zadbaj, aby wykresy miały wspólną oś x.

KOMENTARZ Czerwinski: Wykresy juz maja odpowiednie tytuly i daty na osi x, jako cwiczenie plotuje dwie linie na jednym wykresie

```python
import plotly.graph_objects as go  # Import graph_objects

fig = go.Figure()  # Create an empty figure

fig.add_trace(go.Scatter(x=merged_inner['Data'], y=merged_inner['KGHM'], mode='lines', name='KGHM', yaxis="y"))
fig.add_trace(go.Scatter(x=merged_inner['Data'], y=merged_inner['Miedz'], mode='lines', name='Miedz', yaxis="y2"))

fig.update_layout(
    xaxis_title="Rozwoj na przestrzeni lat",  # X-axis Label
    title='Cena KGHM vs Cena Miedzi',
    title_x=0.5,
    yaxis=dict(  # Styling for the primary y-axis
        title=dict(text="Cena akcji KGHM", font=dict(color="blue")), # Title color
        tickfont=dict(color="blue"),
        range=[0,200]
    ),
    yaxis2=dict(  
        overlaying="y",  # Overlay the secondary y-axis on the primary
        side="right",  # Position the secondary y-axis on the right
        title=dict(text="Cena akcji miedzi", font=dict(color="red")), # Title color
        tickfont=dict(color="red"),
        range=[4000,8000]
    ),
    legend=dict(
        x=0.94
    )
)
fig.show()
```
